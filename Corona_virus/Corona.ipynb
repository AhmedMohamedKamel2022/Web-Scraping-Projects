```python
import requests
from bs4 import BeautifulSoup
url = 'https://dailymedicalinfo.com/coronavirus/'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
list_country = []
countries = soup.find_all('table', attrs = {'id' :'dataTable'})[0].find_all('td')
country = countries[0::5]
# country

for i in country:
    list_country.append(i.text.strip())
print(list_country)
```

```python
list_injuries = []
info_injuries = soup.find_all('table', attrs = {'id' :'dataTable'})[0].find_all('td')
injuries = info_injuries[1::5]
# injuries

for i in injuries:
    list_injuries.append(i.text)
print(list_injuries)
```

```python
list_new_cases = []
info_new_cases = soup.find_all('table', attrs = {'id' :'dataTable'})[0].find_all('td')
new_cases = info_new_cases[2::5]
# new_cases

for i in new_cases:
    list_new_cases.append(i.text)
print(list_new_cases)
```

```python
list_death = []
info_death = soup.find_all('table', attrs = {'id' :'dataTable'})[0].find_all('td')
death = info_death[3::5]
# death

for i in death:
    list_death.append(i.text)
print(list_death)
```

```python
list_recover = []

info_recover = soup.find_all('table',attrs = {'id':'dataTable'})[0].find_all('td')[4::5] 
# info_recover
info_recover = soup.find_all('b')
recover = info_recover
# recover

for i in recover:
    list_recover.append(i.text)
print(list_recover)
```

```python
import csv
from itertools import zip_longest
list_data = [list_country, list_injuries, list_new_cases, list_death, list_recover]
list_unback = zip_longest(*list_data)

with open('Corona.csv', 'w',  encoding="utf-8") as file:
    
    wr = csv.writer(file)
    wr.writerow(['country', 'injuries', 'new_cases', 'death', 'recover'])
    wr.writerows(list_unback)
```
