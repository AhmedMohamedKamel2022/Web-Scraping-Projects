```python
import csv
with open('Jobs.csv', 'w', encoding='utf-8') as file:
    wr= csv.DictWriter(file, fieldnames=['Title', 'Company', 'Address', 'Time', 'Requirements'])
    wr.writeheader()   
    
    next_page = soup.find('div').find_all('button')[3]
    page=0
    while page <688:
        for i in next_page:
            next_link =  f'https://wuzzuf.net/search/jobs/?start={page}'
            print(next_link)

            response = requests.get(next_link)
            soup = BeautifulSoup(response.text, 'html.parser')
            scrap = soup.find_all('div', attrs={'class': 'css-1gatmva e1v1l3u10'})
            for i in scrap:
                name = i.find_all('a')[1].text
                company = i.find_all('a')[2].text.split('-')[0]
                address = i.find_all('span')[0].text
                time = i.find_all('span')[1].text
                req = i.find_all('div')[4].text  
                print({'Title': name, 'Company': company, 'Address': address, 'Time': time, 'Requirements': req})
                wr.writerow({'Title': name, 'Company': company, 'Address': address, 'Time': time, 'Requirements': req})
            page+=1

        
```
