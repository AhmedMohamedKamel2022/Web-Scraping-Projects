```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://wuzzuf.net/search/jobs/?q=python+develober&a=navbl'
resopse = requests.get(url)
# print(resopse.text)
```

```python
soup = BeautifulSoup(resopse.text,'html.parser')
```

```python
job_title = soup.find_all('a', attrs={'target' : '_blank', 'class' : 'css-o171kl'})
# job_title

location = soup.find_all('span', attrs={'class' : 'css-5wys0k'})
# location

companies_name = soup.find_all('a',attrs = {'class' : 'css-17s97q8'})
# companies_name

late_posted_time = soup.find_all('div', attrs = {'class' : 'css-do6t5g'})
# late_posted_time

early_posted_time = soup.find_all('div', attrs = {'class' : 'css-4c4ojb'})
# early_posted_time
```

```python
l_job = []
for i in job_title:
    l_job.append(i.text)
print(l_job,'\n')


l_location = []
for i in location:
    l_location.append(i.text)
print(l_location,'\n')


l_companies_name = []
for i in companies_name:
    l_companies_name.append(i.text)
print(l_companies_name,'\n')


l_late_posted_time = []
for i in late_posted_time:
    l_late_posted_time.append(i.text)
print(l_late_posted_time,'\n')


l_early_posted_time = []
for i in early_posted_time:
    l_early_posted_time.append(i.text)
print(l_early_posted_time)

```

```python
import csv
from itertools import zip_longest

file_list = [ l_job, l_location, l_companies_name, l_late_posted_time, l_early_posted_time ]
exported = zip_longest(*file_list)

with open('jobs.csv', 'w') as file:
    wr = csv.writer(file)
    wr.writerow(['job_title', 'location', 'companies_name', 'late_posted_time', 'early_posted_time'])
    wr.writerows(exported)
```
