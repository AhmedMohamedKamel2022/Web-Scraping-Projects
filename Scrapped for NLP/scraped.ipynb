```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://m.marefa.org/1000_%D8%B3%D8%A4%D8%A7%D9%84_%D9%81%D9%8A_%D8%A7%D9%84%D8%AA%D8%A7%D8%B1%D9%8A%D8%AE_%D8%A7%D9%84%D8%A5%D8%B3%D9%84%D8%A7%D9%85%D9%8A'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# soup
```

```python
scrap = soup.find_all('pre')
# scrap
```

```python
data_1 = []
for i in scrap:
    i = i.text
    data_1.append(i.strip().split('\n'))
print(data_1)
```

```python
scrap_2 = soup.find_all('p')[48:-2]
scrap_2
```

```python
data_2 = []
for i in scrap_2:
    i = i.text
    data_2.append(i.strip().split('\n'))
print(data_2)
```
