```python
import requests
url = 'https://www.metaweather.com/api/location/1521894/'
response = requests.get(url)
data = response.json()
print(data,'\n')

temp_list = []
for i in data['consolidated_weather']:
    temp_list.append(i['the_temp']) 
print('temp : ', temp_list,'\n')


wind_list = []
for i in data['consolidated_weather']:
    wind_list.append(i['wind_speed'])
print('wind_speed : ', wind_list,'\n')


humidity_list = []
for i in data['consolidated_weather']:
    humidity_list.append(i['humidity']) 
print('humidity : ', humidity_list,'\n')


visibility_list = []
for i in data['consolidated_weather']:
    visibility_list.append(i['visibility']) 
print('visibility : ', visibility_list,'\n')


air_pressure_list = []
for i in data['consolidated_weather']:
    air_pressure_list.append(i['air_pressure']) 
print('air_pressure : ', air_pressure_list,'\n')


import csv
from itertools import zip_longest
from datetime import datetime
    
list_data = [temp_list, wind_list, humidity_list, visibility_list, air_pressure_list]
tuple_unback = zip_longest(*list_data)
    
with open('weather_2_2.csv', 'w') as file:
    header = ['Temp', 'Wind_speed', 'Humidity', 'Visibility', 'Air_pressure']
    wr = csv.writer(file)
    wr.writerow(header)
    wr.writerows(tuple_unback)
    
    times = datetime.now()
    time_now = times.strftime("%d/%m/%Y %H:%M:%S")
    
    output = {'Temp': temp_list, 'Wind': wind_list, 'Humidity': humidity_list, 'Visibility': visibility_list, 'Pressure': air_pressure_list, 'time_now': time_now}
    print(output)
```
