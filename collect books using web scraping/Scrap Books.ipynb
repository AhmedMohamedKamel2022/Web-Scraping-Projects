```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
url = 'https://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

## scrap each category

```python
div = soup.find('div', attrs={'class': 'side_categories'})
# div
ul = div.find_all('ul')[1]
# ul
li = ul.find_all('li')
# li
```

```python
for item in li:
    a = item.find('a')
    category = a.text.strip()
    print(a)
    print(category)
```

## scrap next pages

```python
nexts = soup.find_all('li',attrs={'class':'next'})
nexts
```

```python
page = 1
while page <=50:
    for i in nexts:
        a = i.find('a')
        link_next = a.attrs['href']
#         print(a)
#         print(link_next)
        link = f'catalogue/page-{page}.html'
        print(link)
        page+=1
```

## scrap name of book

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    name = i.find('h3').text
    print(name)
```

## scrap price of book

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    price = i.find('p', attrs={'class': 'price_color'}).text[2:]
    print(price)
```

## scrap rate of book

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    rate = i.find('p').get('class')[1]
    print(rate)
```
