```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
url = 'https://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
div = soup.find('div', attrs={'class': 'side_categories'})
# div
ul = div.find_all('ul')[1]
# ul
li = ul.find_all('li')
# li
```

```python
for item in li:
    a = item.find('a')
    category = a.text.strip()
    print(a)
    print(category)
```

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    name = i.find('h3').text
    print(name)
```

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    price = i.find('p', attrs={'class': 'price_color'}).text
    print(price)
```

```python
scrap = soup.find_all('article', attrs={'class': 'product_pod'})
for i in scrap:
    rate = i.find('p').get('class')[1]
    print(rate)
```

## Complete code

```python
for item in li:
    
    a = item.find('a')
    category = a.text.strip()
    category_link =  a.attrs['href'] 
    
    link = url + category_link
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    scrap = soup.find_all('article', attrs={'class': 'product_pod'})
    
    for i in scrap:
        
        name = i.find('h3').text
        price = i.find('p', attrs={'class': 'price_color'}).text
        rate = i.find('p').get('class')[1]
        print({'Title of book': name, 'Price of book': price, 'Rate of book': rate, 'Category of book': category})
```

```python
import csv
with open('Books.csv', 'w', encoding='utf-8') as file:
    wr = csv.DictWriter(file, fieldnames=['Title', 'Price', 'Rate', 'Category'])
    wr.writeheader()
    rates = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}

    for item in li:
        a = item.find('a')
        category = a.text.strip()
        category_link =  a.attrs['href'] 

        link = url + category_link
        response = requests.get(link)
        soup = BeautifulSoup(response.text, 'html.parser')

        scrap = soup.find_all('article', attrs={'class': 'product_pod'})
        for i in scrap:
            name = i.find('h3').text
            price = i.find('p', attrs={'class': 'price_color'}).text
            rate = i.find('p').get('class')[1]
            rate = rates[rate]
            
            wr.writerow({'Title': name, 'Price': price, 'Rate': rate, 'Category': category})
```
